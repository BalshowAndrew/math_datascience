### Задание 1

В магазин привезли устройства с 3-х разных предприятий.

Соотношение устройств следующее: 20% - продукция первого предприятия, 30% - продукция второго предприятия, 50% - продукция третьего предприятия; далее, 10% продукции первого предприятия высшего сорта, на втором предприятии - 5% и на третьем - 20% продукции высшего сорта.

Найти вероятность того, что случайно купленная новая продукция окажется высшего сорта.

### Решение

```python
# Вероятность совершения покупки продукции определенного предприятия
pH1 = 0.2 # первое предприятие
pH2 = 0.3 # второе предприятие
pH3 = 0.5 # третье предприятие
```

```python
# Вероятности покупки продукции высшего сорта при совершении покупки продукции каждого из предприятий
pAH1 = 0.1
pAH2 = 0.05
pAH3 = 0.2
```

```python
# Расчет полной вероятности:
pA = (pAH1 * pH1) + (pAH2 * pH2) + (pAH3 * pH3)
print(f"Вероятность того, что случайно купленная новая продукция окажется высшего сорта равна {pA}")
```

### Задание 2

Пусть брошены 3 уравновешенные монеты.

Рассмотрим 3 события:

A1 - монеты 1 и 2 упали одной и той же стороной;

A2 - монеты 2 и 3 упали одной и той же стороной;

A3 - монеты 1 и 3 упали одной и той же стороной.

Покажите, почему эти 3 события (A1, A2, A3) являются попарно независимыми, но не являются независимыми в совокупности.

### Решение

- Выпадения какой-либо из сторон каждой из трех монет явлются попарно независимыми событиями, так как вероятность наступления любого из них не зависит от появления остальных событий.
- В совокупности же вероятность выпадения какой либо из сторон монеты при многократных повторениях зависит от серии событий, предшествующих данному.

### Задание 3

Изучите 3 распределения из библиотеки scipy: Нормальное, Экспоненциальное, Стьюдента. Используя функцию плотности распредления (pdf), постройте пять графиков плотностей для каждого распределения при разных параметрах. Запишите ваши наблюдения: как и на что влияют параметры у каждого распределения.

Дополнительная информация тут.

### Решение:

import numpy as np 
from scipy import stats
import matplotlib.pyplot as plt

#### Нормальное распределение

```python
x = np.linspace(-4, 4, 200)
```

```python
pdf_norm = stats.norm.pdf(x, loc=1, scale=0.5)
```

```python
plt.plot(x, pdf_norm)
plt.xlabel('x-values')
plt.ylabel('pdf_norm_values')
plt.title('График плотности нормального распределения')
plt.show()
```

Параметр **loc** смещает вершину графика, а переметр **scale** меняет его кривизну

#### Экспоненциальное распределение

```python
fig, ax = plt.subplots(1, 1)
x = np.linspace(stats.expon.ppf(0.01),
                stats.expon.ppf(0.99), 100)
ax.plot(x, stats.expon.pdf(x, loc=0.1, scale=1),
       'r-', lw=5, alpha=0.6, label='expon pdf')
plt.xlabel('x-values')
plt.ylabel('pdf_expon_values')
plt.title('График плотности экспоненциального распределения')
plt.show()
```

#### Распределение Стьюдента

```python
fig, ax = plt.subplots(1, 1)

df=2.5
x = np.linspace(stats.t.ppf(0.01, df),
                stats.t.ppf(0.99, df), 100)
ax.plot(x, stats.t.pdf(x, df, loc=1, scale=3),
       'r-', lw=5, alpha=0.6, label='t pdf')
plt.xlabel('x-values')
plt.ylabel('pdf_t_values')
plt.title('График плотности распределения Стьюдента')
plt.show()
```

Параметр **loc** смещает вершину графика, а переметр **scale** меняет его кривизну
